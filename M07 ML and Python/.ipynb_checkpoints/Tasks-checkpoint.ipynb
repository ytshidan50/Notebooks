## 1. Accept a char input from the user and display it on the console.

```python
charNotEntered=True
while(charNotEntered):
    print("Hi")
    ch=str(input("Enter a character: "))
    if len(ch)==1:
        charNotEntered=False
        print(f"Your entered character '{ch}' successfully verified.")
    else:
        print("Invalid character. ",end='')
```

6. Using the ‘switch case’ write a program to accept an input number from the user and output the day as follows.

| **Input** | **Output** |
| --- | --- |
| 1 | Sunday |
| 2 | Monday |
| 3 | Tuesday |
| 4 | Wednesday |
| 5 | Thursday |
| 6 | Friday |
| 7 | Saturday |
| Any other input | Invalid Entry |

```python
d=int(input("Enter a number to find the corresponding day: "))
match d:
    case 1:
        print("Sunday")
    case 2:
        print("Monday")
    case 3:
        print("Tuesday")
    case 4:
        print("Wednesday")
    case 5:
        print("Thursday")
    case 6:
        print("Friday")
    case 7:
        print("Saturday")
    case _:
        print("Invalid Entry")
```

## 9. Write a program to create a copy of array and add an element to copied array . show both arrays.
    a. Program should accept an array from the user, swap the values of two arrays and display it on the console

Eg: 

**Output**: Enter the size of Array 1

**Input**: 5

**Output**: Enter the values of Array 1

Array 1: 10, 20, 30, 40, 50

Array 2 : Copy of Array 1

Array 2 : 10, 20, 30, 40, 50 + add a element

**Output :** Array 1 **[** 10, 20, 30, 40, 50 **]**

**Output :** Array 2 **[** 10, 20, 30, 40, 50, 60, 70 **]**

```python
a1c=int(input("Enter the size of Array 1: "))
a1=[input(f"Enter the value for pos {i+1}: ") for i in range(a1c)]
print("Array 1: ",end='')
[print(x,end=' ') for x in a1]
print()
a2=a1.copy()
print("Array 2 (Copy): ",end='')
[print(x,end=' ') for x in a2]
c=int(input("Enter the count of numbers which you want to add additionally into the copied array (Array 2): "))
for i in range(c):
    a2.append(input(f"Enter the {i+1}-value for pos {len(a2)+1}: "))
print("Array 1: ",end='')
[print(x,end=' ') for x in a1]
print()
print("Array 2: ",end='')
[print(x,end=' ') for x in a2]
print()
```

## 10. Write a program to sort an array in descending order without sort() and sorted()
    a. Program should accept and array, sort the array values in descending order and display it

Eg: 

**Output**: Enter the size of an array

**Input**: 5

**Output**: Enter the values of array

**Input**: 20, 10, 50, 30, 40

**Output**: Sorted array: 50, 40, 30, 20, 10

```python
c=int(input("Enter the length of the array: "))
a=[input(f"Enter the value for pos {i+1}: ") for i in range(c)]
for i in range(c):
    for j in range(i+1,c):
        if a[j]>a[i]:
            temp=a[i]
            a[i]=a[j]
            a[j]=temp
print("The sorted array: ",end='')
[print(a[i], end=', ' if i!=c-1 else '') for i in range(c)]
print()
```

## 11. Write a program to identify whether a string is a palindrome or not without using reverse(), slicing
    a. A string is a palindrome if it reads the same backward or forward eg: MALAYALAM
    b. Program should accept a string and display whether the string is a palindrome or not

Eg: 

**Output**: Enter a string

**Input**: MALAYALAM

**Output**: Entered string is a palindrome

Eg 2: 

**Output**: Enter a string

**Input**: HELLO

**Output**: Entered string is not a palindrome

```python
s=str(input("Enter the string to check: "))
pallindrome=True
for i in range(int(len(s)-1/2)):
    if s[i]!=s[len(s)-1-i]:
        pallindrome=False
        break
if pallindrome:
    print("The string is a pallindrome.")
else:
    print("The entered string is not a pallindrome!")
```
